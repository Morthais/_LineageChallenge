# DATA IMPORTS

```python
# imports
import pandas as pd
import altair as alt
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in all the dataframes for cross-analysis
shipwreck = pd.read_csv("sample_geospatial_shipwrecks.csv")
planets = pd.read_csv("sample_guides_planets.csv")
movies = pd.read_csv("movies.csv")
inspections = pd.read_csv("inspections.csv")
sales = pd.read_csv("sales.csv")
weather = pd.read_csv("weather.csv")
listings = pd.read_csv("listingsAndReviews.csv")
grades = pd.read_csv("grades.csv")
```

# DATA CLEANUP

```python
shipwreck_clean = shipwreck.drop_duplicates()
planets_clean = planets.drop_duplicates()
movies_clean = movies.drop_duplicates()
inspections_clean = inspections.drop_duplicates()
sales_clean = sales.drop_duplicates()
weather_clean = weather.drop_duplicates()
listings_clean = listings.drop_duplicates()
grades_clean = grades.drop_duplicates()

shipwreck_clean = shipwreck.replace(np.nan, 0)
planets_clean = planets.replace(np.nan, 0)
movies_clean = movies.replace(np.nan, 0)
inspections_clean = inspections.replace(np.nan, 0)
sales_clean = sales.replace(np.nan, 0)
weather_clean = weather.replace(np.nan, 0)
listings_clean = listings.replace(np.nan, 0)
grades_clean = grades.replace(np.nan, 0)
```

### Planets Correlations

```python
# https://www.vertica.com/blog/in-database-machine-learning-2-calculate-a-correlation-matrix-a-data-exploration-post/ 
# https://www.geeksforgeeks.org/how-to-create-a-correlation-matrix-using-pandas/ 
planets_clean.corr()
```

```python
# distance from the sun and max surface temperature are negatively correlated
# this mean that the further a planet is from the sun, the colder its maximum surface temperature
plt.figure(figsize=(9,5))
plt.xlabel('Distance from Sun')
plt.ylabel('Max Surface Temperature')
plt.title('Planet Surface Temperature Decreases the Farther the Planet is from the Sun')
sns.lineplot(data=planets_clean, x='orderFromSun', y='surfaceTemperatureC.max');
```

```python
# distance from the sun and whether a planet has rings are positively correlated
# this means the further a planet is from the sun the more likely it is to have rings
plt.figure(figsize=(9,5))
plt.xlabel('Distance from Sun')
plt.ylabel('Planet has Rings')
plt.title('Planets Farther than the Sun in Our Solar System Have Rings')
sns.lineplot(data=planets_clean, x='orderFromSun', y='hasRings')
```

### Movies Correlations

```python
movies_clean.info()
```

```python
# in case you do not care about visualizations and just want to know what the data says...
# this measures the correlation rating between columns in the data
# closer to 1 means higher correlation and closer to 0 means lower correlation
movies_clean.corr()
```

```python
# the number of award nominations is positively correlated with the number of award wins
# this means the more times a movie is nominated for an award the more likely they will win
# therefore, a director has nothing to lose from nominating their own movie for an award
plt.figure(figsize=(9,5))
plt.xlabel('Total Awards Won')
plt.ylabel('Total Award Nominations')
plt.title('Movie Award Nominations Increase the Likelihood of Award Wins')
sns.scatterplot(data=movies_clean, x='awards.wins', y='awards.nominations');
```

```python
# the number of imdb votes is somewhat positively correlated with the number of award wins
# this means the number of imdb votes has a smaller impact on whether a movie wins an award
# therefore, it would be good to have lots of imdb votes to increase chances of an award win
plt.figure(figsize=(9,5))
plt.xlabel('Total IMDB Votes')
plt.ylabel('Total Award Nominations')
plt.title('Movie Award Nominations More Likely with More IMDB Votes')
sns.scatterplot(data=movies_clean, x='imdb.votes', y='awards.nominations');
```
